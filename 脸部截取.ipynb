```python
import dlib
import cv2
import os
# 加载并初始化检测器

path=r"E:\JupyterWorkspace\DFDC\test_data\\"  #待读取的文件夹
path_list=os.listdir(path)

for filename in path_list:
    if filename.endswith(('mp4')):
        detector = dlib.get_frontal_face_detector()
        camera = cv2.VideoCapture(path+filename)
        if not camera.isOpened():
            print("cannot open camear")
            exit(0)
        j=0
        while True:
            ret, frame = camera.read()
            if j is 1:
                break
            frame_new = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            # 检测脸部
            dets = detector(frame_new, 1)
            print("Number of faces detected: {}".format(len(dets)))
            # 查找脸部位置
            for i, face in enumerate(dets):
                print("Detection {}: Left: {} Top: {} Right: {} Bottom: {} ".format(i, face.left(), face.top(), face.right(), face.bottom()))
                #保存脸部图片,注意这里+ - 是根据dlib对脸部的定位来的，我这里选的50很适合我这个视频
                img1=frame[face.top()-50:face.bottom()+50,face.left()-50:face.right()+50]
                crop_size = (256,256)
                img1 = cv2.resize(img1, crop_size, interpolation = cv2.INTER_CUBIC)
                cv2.imwrite(r'E:\JupyterWorkspace\DFDC\test_pic\\'+filename+'_{}.jpg'.format(j), img1)  
                j=j+1
                if j== 1:
                    break
            cv2.imshow("Camera", frame)
            key = cv2.waitKey(1)
            if key == 27:
                break
        cv2.destroyAllWindows()
```
